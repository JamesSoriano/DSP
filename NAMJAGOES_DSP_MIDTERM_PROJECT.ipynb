```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files

uploaded = files.upload()
data = pd.read_excel(list(uploaded.keys())[0])

data['TIME'] = pd.to_datetime(data['TIME'], format='%H:%M:%S')
data['DATE'] = pd.to_datetime(data['DATE'], format='%m-%d-%Y')
data['DATETIME'] = data['DATE'] + pd.to_timedelta(data['TIME'].dt.strftime('%H:%M:%S'))

aveT = data['TEMPERATURE'].mean()
aveH = data['HUMIDITY'].mean()
dp = len(data)

plt.figure(figsize=(14, 6))
plt.scatter(data['DATETIME'], data['TEMPERATURE'], c='blue', label=f'Temperature C (Avg: {aveT:.2f} C)')
plt.scatter(data['DATETIME'], data['HUMIDITY'], c='red', label=f'Humidity %RH (Avg: {aveH:.2f} %RH)')
plt.xlabel('Date and Time')
plt.ylabel('Temperature and Humidity')
plt.title(f'Temperature and Humidity Scatter Plot ({dp} data points)')

plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data['TIME'] = pd.to_datetime(data['TIME'], format='%H:%M:%S')
data['DATE'] = pd.to_datetime(data['DATE'], format='%m-%d-%Y')
data['DATETIME'] = data['DATE'] + pd.to_timedelta(data['TIME'].dt.strftime('%H:%M:%S'))

aveT = data['TEMPERATURE'].mean()
aveH = data['HUMIDITY'].mean()
dp = len(data)

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 8), sharex=True)

ax1.plot(data['DATETIME'], data['TEMPERATURE'], color='blue', label=f'Temperature C (Avg: {aveT:.2f} C)')
ax1.set_ylabel('Temperature (C)')
ax1.set_title(f'Temperature and Humidity Periodicity Plot ({dp} data points)')
ax1.legend()

ax2.plot(data['DATETIME'], data['HUMIDITY'], color='red', label=f'Humidity %RH (Avg: {aveH:.2f} %RH)')
ax2.set_xlabel('Date and Time')
ax2.set_ylabel('Humidity (%RH)')
ax2.legend()

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

**Morning and Evening Filtered Data**

```python
import pandas as pd
import matplotlib.pyplot as plt

data['TIME'] = pd.to_datetime(data['TIME'], format='%H:%M:%S')
data['DATE'] = pd.to_datetime(data['DATE'], format='%m-%d-%Y')
data['DATETIME'] = data['DATE'] + pd.to_timedelta(data['TIME'].dt.strftime('%H:%M:%S'))

morning = data[data['DATETIME'].dt.hour < 12]
evening = data[data['DATETIME'].dt.hour >= 12]

aveT_am = morning['TEMPERATURE'].mean()
aveH_am = morning['HUMIDITY'].mean()
aveT_pm = evening['TEMPERATURE'].mean()
aveH_pm = evening['HUMIDITY'].mean()

plt.figure(figsize=(14, 6))
plt.scatter(morning['DATETIME'], morning['TEMPERATURE'], c='blue', label=f'Morning Temperature C (Avg: {aveT_am:.2f} C)')
plt.scatter(morning['DATETIME'], morning['HUMIDITY'], c='red', label=f'Morning Humidity %RH (Avg: {aveH_am:.2f} %RH)')

plt.scatter(evening['DATETIME'], evening['TEMPERATURE'], c='green', label=f'Evening Temperature C (Avg: {aveT_pm:.2f} C)')
plt.scatter(evening['DATETIME'], evening['HUMIDITY'], c='orange', label=f'Evening Humidity %RH (Avg: {aveH_pm:.2f} %RH)')

plt.xlabel('Date and Time')
plt.ylabel('Temperature and Humidity')
plt.title(f'Temperature and Humidity Scatter Plot')

plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

**Correlation between temperature and humidity**

```python
import pandas as pd
import matplotlib.pyplot as plt

data['TIME'] = pd.to_datetime(data['TIME'], format='%H:%M:%S')
data['DATE'] = pd.to_datetime(data['DATE'], format='%m-%d-%Y')
data['DATETIME'] = data['DATE'] + pd.to_timedelta(data['TIME'].dt.strftime('%H:%M:%S'))

correlation = data['TEMPERATURE'].corr(data['HUMIDITY'])

plt.figure(figsize=(8, 6))
plt.scatter(data['TEMPERATURE'], data['HUMIDITY'], c='blue', label=f'Temperature vs Humidity (Correlation: {correlation:.2f})')
plt.xlabel('Temperature (C)')
plt.ylabel('Humidity (%RH)')
plt.title('Temperature vs Humidity Scatter Plot')

plt.legend()
plt.tight_layout()
plt.show()
```
